```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.plot(x, y, color='blue')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D']
sales = [150, 230, 180, 200]

plt.bar(products, sales, color='orange')
plt.title('Bar Chart Example')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

height = [150, 160, 170, 180, 190]
weight = [50, 55, 65, 75, 85]

plt.scatter(height, weight, color='green')
plt.title('Scatter Plot Example')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
import matplotlib.pyplot as plt

activities = ['Study', 'Sleep', 'Eat', 'Play']
time = [8, 7, 5, 4]

plt.pie(time, labels=activities )
plt.title('Pie Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [45, 67, 56, 78, 89, 90, 72, 60, 83, 95]

plt.hist(marks, bins=[40, 50, 60, 70, 80, 90, 100], color='skyblue', edgecolor='black')
plt.title('Histogram - Students Marks Distribution')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.show()
```
